```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming 'uni' is your original DataFrame and 'cols' are the column names
scaler = StandardScaler()
scaled_uni_df = pd.DataFrame(scaler.fit_transform(uni), columns=cols)

# Output the scaled DataFrame
scaled_uni_df
```

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state = 0)
clusters_new.fit(scaled_uni_df)
```

#### finding optimal k value using elow plot


```python
wcss = []
for i in range(1,20):
    Kmeans = KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(scaled_uni_df)
    wcss.append(Kmeans.inertia)
print(wcss)
plt.plot(range(1,20),wcss)
plt.title('Elbow Method')
plt.xlabel('Number f clusters')
plt.title('WCSS')
plt.show()

    
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Initialize an empty list to store the WCSS values
wcss = []

# Loop to fit KMeans for different numbers of clusters
for i in range(1, 20):
    Kmeans = KMeans(n_clusters=i, random_state=0)
    Kmeans.fit(scaled_uni_df)
    wcss.append(Kmeans.inertia_)

# Print the WCSS values
print(wcss)

# Plot the Elbow Method
plt.plot(range(1, 20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
