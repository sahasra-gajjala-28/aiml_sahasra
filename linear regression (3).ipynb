```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
model1.summary()
```

```python
# plot the scatter plot and oevrlay the fitted straight line using matpltlib
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s=30)
b0 =13.84
b1 = 1.33
#predicate response vector
y_hat = b0+ b1*x
#plotting the regression line
plt.plot(x, y_hat, color = "g")
# putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
sns.regplot(x="daily", y="sunday", data=data)
plt.xlim([0,1250])
plt.show()
```

```python
#predict sunday sales for 200 300 & 1500 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data["daily"])
pred
```

```python
data["y_hat"] = pred
data
```

```python
data["residuals"] = data["sunday"]-data["y_hat"]
data
```

```python
mse = np.mean((data["daily"]-data["y_hat"])**2)
rmse =np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```
